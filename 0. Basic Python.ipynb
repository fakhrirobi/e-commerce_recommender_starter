# Basic Python

This is a variables

```python
a = 1
b = 1.5
c = [1, 2, 3, 4]
d = {'a': 3,
     'b': 4}
```

Print using a `print()` syntax

```python
print(a)
print(b)
print(c)
print(d)
```

Operate the variables

```python
sum_ = a + b
sub_ = a - b
mul_ = a * b
div_ = a / b
```

```python
print(sum_, sub_, mul_, div_)
```

Accessing variables

```python
print(c[1])
print(d['b'])
```

Looping

```python
for num in c:
    print(num)
```

```python
for key, val in d.items():
    print(key, ':', val)
```

Function

```python
def sum(a, b):
    """Function to do summation between a & b"""
    return a + b
```

```python
c = sum(a=5, b=3)
print(c)
```
